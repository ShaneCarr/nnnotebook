```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the symbolic variable x and a small increment h
x, h = sp.symbols('x h')

# Define the function f(x) = x^3
f = x**3

# Step 1: Find f(x+h)
f_x_plus_h = f.subs(x, x + h)

# Step 2: Calculate the difference quotient (f(x+h) - f(x))/h
difference_quotient = (f_x_plus_h - f) / h

# Step 3: Simplify the difference quotient
simplified_difference_quotient = sp.simplify(difference_quotient)

# Step 4: Take the limit as h approaches 0 to find the derivative
derivative = sp.limit(simplified_difference_quotient, h, 0)

# Create a plot to visualize the process
plt.figure(figsize=(12, 6))

# Plot the original function
plt.subplot(231)
x_vals = np.linspace(-5, 5, 400)
f_vals = [f.subs(x, val) for val in x_vals]
plt.plot(x_vals, f_vals, label='f(x) = x^3', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Step 1: Original Function')
plt.legend()

# Plot f(x+h)
plt.subplot(232)
f_x_plus_h_vals = [f_x_plus_h.subs({x: val, h: 0.1}) for val in x_vals]
plt.plot(x_vals, f_x_plus_h_vals, label='f(x+h) with h=0.1', color='green')
plt.xlabel('x')
plt.ylabel('f(x+h)')
plt.title('Step 2: f(x+h)')
plt.legend()

# Plot the difference quotient
plt.subplot(233)
difference_quotient_vals = [difference_quotient.subs({x: val, h: 0.1}) for val in x_vals]
plt.plot(x_vals, difference_quotient_vals, label='Difference Quotient', color='orange')
plt.xlabel('x')
plt.ylabel('(f(x+h) - f(x))/h')
plt.title('Step 3: Difference Quotient')
plt.legend()

# Plot the simplified difference quotient
plt.subplot(234)
simplified_difference_quotient_vals = [simplified_difference_quotient.subs({x: val, h: 0.1}) for val in x_vals]
plt.plot(x_vals, simplified_difference_quotient_vals, label='Simplified Diff. Quotient', color='purple')
plt.xlabel('x')
plt.ylabel('Simplified Diff. Quotient')
plt.title('Step 4: Simplified Diff. Quotient')
plt.legend()

# Plot the derivative
plt.subplot(235)
derivative_vals = [derivative.subs(x, val) for val in x_vals]
plt.plot(x_vals, derivative_vals, label="f'(x)", color='red')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Step 5: Derivative')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import sympy as sp

# Define the symbolic variable x and a small increment h
x, h = sp.symbols('x h')

# Define the function f(x) = x^3
f = x**3

# Step 1: Find f(x+h)
f_x_plus_h = f.subs(x, x + h)

# Step 2: Expand f(x+h)
expanded_f_x_plus_h = sp.expand(f_x_plus_h)

# Step 3: Calculate f(x+h) - f(x)
difference = expanded_f_x_plus_h - f

# Step 4: Divide by h
difference_quotient = difference / h

# Step 5: Simplify the difference quotient
simplified_difference_quotient = sp.simplify(difference_quotient)

# Step 6: Take the limit as h approaches 0 to find the derivative
derivative = sp.limit(simplified_difference_quotient, h, 0)

# Print the symbolic expressions at each step
print("Step 1: f(x+h) = ", f_x_plus_h)
print("\nStep 2: Expanded f(x+h) = ", expanded_f_x_plus_h)
print("\nStep 3: f(x+h) - f(x) = ", difference)
print("\nStep 4: Difference Quotient = ", difference_quotient)
print("\nStep 5: Simplified Difference Quotient = ", simplified_difference_quotient)
print("\nStep 6: Derivative (as h approaches 0) = ", derivative)
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define the symbolic variable x
x = sp.symbols('x')

# Define the function f(x) = 2x
f = 2 * x

# Step 1: Compute the indefinite integral
integral = sp.integrate(f, x)

# Create a plot to visualize the process
plt.figure(figsize=(10, 6))

# Plot the original function
x_vals = np.linspace(0, 5, 400)
f_vals = [f.subs(x, val) for val in x_vals]
plt.plot(x_vals, f_vals, label='f(x) = 2x', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Original Function')
plt.legend()

# Plot the integral
integral_func = sp.lambdify(x, integral, 'numpy')
integral_vals = [integral_func(val) for val in x_vals]
plt.figure(figsize=(10, 6))
plt.plot(x_vals, integral_vals, label='Integral of f(x)', color='green')
plt.xlabel('x')
plt.ylabel('Integral of f(x)')
plt.title('Integral of Original Function')
plt.legend()

plt.tight_layout()
plt.show()

print("The integral of f(x) = 2x is:", integral)
```

```python
import sympy as sp
print("Proof of antiderative")

# Define symbolic variables and constants
x, a, b, C = sp.symbols('x a b C')

# Define a function f(x)
f = sp.sin(x)

# Find an antiderivative F(x) of f(x)
F = sp.integrate(f, x) + C

# Define the definite integral from a to b
definite_integral = sp.integrate(f, (x, a, b))

# Apply the Fundamental Theorem of Calculus
result = F.subs(x, b) - F.subs(x, a)

# Display the results
print("Function f(x):", f)
print("\nAntiderivative F(x) of f(x):", F)
print("\nDefinite Integral [a, b] f(x) dx:", definite_integral)
print("\nResult using the Fundamental Theorem of Calculus:", result)
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define symbolic variables and constants
x, a, b, C = sp.symbols('x a b C')

# Define the function f(x) = 2x
f = 2 * x

# Define the limits of integration
a_val = 1
b_val = 4

# Define the number of rectangles for visualization
num_rectangles = 10

# Define the width of each rectangle (delta x)
delta_x = (b_val - a_val) / num_rectangles

# Calculate the x-values for the rectangles
x_values = np.linspace(a_val, b_val, num_rectangles)

# Calculate the corresponding heights of the rectangles (2x)
rectangle_heights = [f.subs(x, val) for val in x_values]

# Calculate the area of each rectangle
rectangle_areas = [height * delta_x for height in rectangle_heights]

# Calculate the total area by summing the areas of all rectangles
total_area = sum(rectangle_areas)

# Create a plot to visualize the rectangles and the curve
plt.figure(figsize=(10, 6))

# Plot the curve (2x)
x_vals = np.linspace(a_val - 1, b_val + 1, 400)
f_vals = [f.subs(x, val) for val in x_vals]
plt.plot(x_vals, f_vals, label='f(x) = 2x', color='blue')

# Plot the rectangles
for i in range(num_rectangles):
    x_rect = [x_values[i], x_values[i] + delta_x, x_values[i] + delta_x, x_values[i], x_values[i]]
    y_rect = [0, 0, rectangle_heights[i], rectangle_heights[i], 0]
    plt.fill(x_rect, y_rect, color='gray', alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualization of Accumulating Area')
plt.legend()

plt.tight_layout()
plt.show()

print("Total Area (Integral) from", a_val, "to", b_val, ":", total_area)
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Define symbolic variables and constants
x, a, b, C = sp.symbols('x a b C')

# Define the function f(x) = 2x (you can change this function)
f = 2 * x

# Define the limits of integration
a_val = 1
b_val = 4

# Define the number of rectangles for visualization
num_rectangles = 10

# Define the width of each rectangle (delta x)
delta_x = (b_val - a_val) / num_rectangles

# Calculate the x-values for the rectangles
x_values = np.linspace(a_val, b_val, num_rectangles)

# Calculate the corresponding heights of the rectangles (2x)
rectangle_heights = [f.subs(x, val) for val in x_values]

# Create a symbolic variable for delta x
delta_x_sym = sp.symbols('delta_x')

# Calculate the area of each rectangle symbolically
rectangle_areas = [f.subs(x, val) * delta_x_sym for val in x_values]

# Calculate the total area (Riemann sum)
total_area = sp.Sum(rectangle_areas, (x, a, b)).doit()

# Create a plot to visualize the rectangles and the curve
plt.figure(figsize=(10, 6))

# Plot the curve (2x)
x_vals = np.linspace(a_val - 1, b_val + 1, 400)
f_vals = [f.subs(x, val) for val in x_vals]
plt.plot(x_vals, f_vals, label='f(x) = 2x', color='blue')

# Plot the rectangles
for i in range(num_rectangles):
    x_rect = [x_values[i], x_values[i] + delta_x, x_values[i] + delta_x, x_values[i], x_values[i]]
    y_rect = [0, 0, rectangle_heights[i], rectangle_heights[i], 0]
    plt.fill(x_rect, y_rect, color='gray', alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Visualization of Accumulating Area')
plt.legend()

plt.tight_layout()
plt.show()

# Display the results
print("Total Area (Riemann Sum) from", a_val, "to", b_val, ":", total_area)
```
